```python
import numpy as np
import pandas as pd

```

```python
df = pd.read_csv('/content/Florida_Alumni_mails_data.csv')
```

```python
First_name  = df['EMAIL'].str.split(',', expand=True)
df['First_Name'] = First_name
df
```

```python
df["First_Name"] = df["First_Name"].str.replace("@ufl.edu", "")
df["First_Name"] = df["First_Name"].str.replace("@peds.ufl.edu", "")
df["First_Name"] = df["First_Name"].str.replace(".", "")
df
```

```python
clean_rows = df.drop_duplicates()
clean_rows
```

```python
clean_rows.to_csv('/content/Florida_Alumni_Data_Final.csv')
```

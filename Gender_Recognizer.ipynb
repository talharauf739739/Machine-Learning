```python
import pandas as pd
import gender_guesser.detector as gender

# Load the data
data = pd.read_excel('/content/May_Medium.xlsx')

# Initialize the gender detector
d = gender.Detector()

# Function to get gender from the first word of the fullname
def get_first_name_gender(fullname):
    if isinstance(fullname, str):
        first_name = fullname.split()[0]  # Get the first word
        return d.get_gender(first_name)   # Predict gender
    else:
        return 'unknown'  # Return 'unknown' if the fullname is not a string

# Process data in batches of 200
batch_size = 200
num_batches = len(data) // batch_size + (1 if len(data) % batch_size != 0 else 0)

# Initialize an empty DataFrame to store the results
result_data = pd.DataFrame()

for i in range(num_batches):
    start_idx = i * batch_size
    end_idx = min((i + 1) * batch_size, len(data))
    batch = data.iloc[start_idx:end_idx].copy()  # Get the current batch

    # Apply the function to the 'fullname' column for the current batch
    batch['gender'] = batch['fullname'].apply(get_first_name_gender)

    # Append the processed batch to the result DataFrame
    result_data = pd.concat([result_data, batch], ignore_index=True)

# Retry processing 'unknown' genders up to 5 times
max_retries = 5
for attempt in range(max_retries):
    # Find rows with 'unknown' gender
    unknown_genders = result_data[result_data['gender'] == 'unknown']
    if unknown_genders.empty:
        break  # Exit if no 'unknown' genders are found

    # Reprocess 'unknown' genders
    for idx, row in unknown_genders.iterrows():
        result_data.at[idx, 'gender'] = get_first_name_gender(row['fullname'])

print(result_data)
result_data.to_excel('/content/May_google.xlsx', index=False)
```

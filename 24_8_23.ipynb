```python
import pandas as pd

# Load your data into a Pandas DataFrame (assuming it's in a CSV file)
# Replace 'your_data.csv' with the path to your data file
df = pd.read_excel('/content/b2b_delivered_data.xlsx')
df.head()

```

```python
# Create a dictionary to store the counts of each 'Phone' value
phone_counts = {}

```

```python
# Iterate through the DataFrame
for index, row in df.iterrows():
    phone = row['Phone']
    discount = row['Coupon Discount']

    # Check if the 'Phone' value has been seen before
    if phone in phone_counts:
        # Increment the count for this 'Phone' value
        phone_counts[phone] += 1

        # Check the first discount value for this 'Phone' value
        if phone_counts[phone] == 2:
            if discount == 500:
                df.at[index, 'Coupon Discount'] = '500 X 2'
            else:
                df.at[index, 'Coupon Discount'] = '500 & ' + str(discount)
        elif phone_counts[phone] == 3:
            if discount == 500:
                df.at[index, 'Coupon Discount'] = '500 X 3'
            else:
                df.at[index, 'Coupon Discount'] = '500 & ' + str(discount)
    else:
        # Initialize the count for this 'Phone' value
        phone_counts[phone] = 1

```

```python
# Save the updated DataFrame to a new CSV file

print(df.head())
df.to_excel('/content/b2b_delivered_data_updated.xlsx', index=False)
```

**Rename the Column Name "Car park to car_park"**

```python
input = df.rename({'Car park':'Carpark'}, axis=1)
input.head()
```

**Now Convert Column Town, country, Location and Performance Data into Numbers 
Using  Label Encoder**

```python
from sklearn.preprocessing import LabelEncoder
le = preprocessing.LabelEncoder()
le.fit(input["Performance"])
print(le.transform(input["Performance"]))
le.fit(input["Carpark"])
print(le.transform(input["Carpark"]))
le_Town = LabelEncoder()
le_Country = LabelEncoder()
le_Location = LabelEncoder()
le_Performance = LabelEncoder()
le_Carpark = LabelEncoder()
```

**Now, by use fit and transform method**
The fit method is calculating the mean and variance of each of the features present in our data. The transform method is transforming all the features using the respective mean and variance
**Also make new columns for them**

```python
input['Town_n'] = le_Town.fit_transform(input['Town'])
input['Country_n'] = le_Country.fit_transform(input['Country'])
input['Location_n'] = le_Location.fit_transform(input['Location'])
input['Performance_n'] = le_Performance.fit_transform(input['Performance'])
input['Carpark_n'] = le_Carpark.fit_transform(input['Carpark'])
```

```python
input.head()
```

**Now Drop the unuseful Columns**

```python
numeric_data = input.drop(['Town','Country','Location', 'Performance','Manager name', 'Carpark'],axis='columns')
numeric_data.head()
```

```python
# import all modules
import seaborn as sns
import matplotlib.pyplot as plt

# plotting two histograms on the same axis
plt.hist(numeric_data['Staff'], bins=5, color='red')
plt.hist(numeric_data['Carpark_n'], bins=5,color='blue')
plt.hist(numeric_data['Competition score'], bins=5,color='orange')




plt.show()
```

**Make "Performance_n" Column as Class LABEL**

```python
numeric_data.drop('Performance_n',axis='columns')
```

```python
target = numeric_data['Performance_n']
target.head()
```

**Import the Decision Tree from Sklearn**

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(numeric_data,target)
```

```python
model.score(numeric_data,target)
```

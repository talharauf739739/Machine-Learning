**For Detect Negative Sentences and Negative Words, I have to do Sentiment Analysis**

```python
import pandas as pd
from textblob import TextBlob



# Define a function to handle missing or invalid values
def analyze_sentiment(text):
    if not isinstance(text, str):
        return "Neutral"  # You can choose "Neutral" or any other suitable label
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity == 0:
        return "Neutral"
    else:
        return "Negative"

# Apply sentiment analysis, handling missing values
data['sentiment'] = data['review'].apply(analyze_sentiment)



```

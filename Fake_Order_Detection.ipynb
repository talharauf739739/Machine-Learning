```python
coupon_counts = Issued_Coupon.groupby('Consignee Name')['issued Coupon'].count().reset_index()
coupon_counts.columns = ['Consignee Name', 'Coupon Count']
coupon_counts.head()
#coupon_counts.to_excel('/content/Coupon_Counts_Against_Names.xlsx')
```

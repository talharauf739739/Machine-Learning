```python

import pandas as pd

# Sample DataFrame (replace this with your actual DataFrame)
data = {
    'Result': [0, 1, 1, 2, 0],
    'Address_Validation': ['Yes', 'Yes', 'No', 'No', 'Yes'],
    'Confusion_Matrix': [''] * 5  # Initialize an empty column for the result
}
df = pd.DataFrame(data)

# Apply the conditions using DataFrame.loc
df.loc[(df['Result'] == 0) & (df['Address_Validation'] == 'Yes'), 'Confusion_Matrix'] = 'True Positive'
df.loc[((df['Result'] == 1) | (df['Result'] == 2)) & (df['Address_Validation'] == 'Yes'), 'Confusion_Matrix'] = 'False Positive'
df.loc[((df['Result'] == 1) | (df['Result'] == 2)) & (df['Address_Validation'] == 'No'), 'Confusion_Matrix'] = 'True Negative'
df.loc[(df['Result'] == 0) & (df['Address_Validation'] == 'No'), 'Confusion_Matrix'] = 'False Negative'

print(df)

```

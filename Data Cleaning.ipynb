## Part 1 - Data Preprocessing

### Importing the dataset

```
df = pd.read_csv('/content/Urgent_Care_2 - Ucare_zocdoc.csv')
```

```
import re

# put the special characters in a list
special_char = ['"', '*', '/', '(', ')', ':', '\n','[',']',"'", ',']

special_char_escaped = list(map(re.escape, special_char))

df['Email'] = df['Email'].replace(special_char_escaped, '', regex=True)
```

```
df
```

**Splitting Coloumn by ,**

```
First_name  = df['Email'].str.split(',', expand=True)
df['First_Name'] = First_name
df

```

**Replacement @ data with ""**

```
df["First_Name"] = df["First_Name"].str.replace("@gmail.com", "")
df["First_Name"] = df["First_Name"].str.replace("@aol.com", "")
df["First_Name"] = df["First_Name"].str.replace(".", "")
df
```

```
df.to_csv('/content/Ucare_ZOCDOCs_general.csv')
```
